```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3), (3, 1), (1, "a")])

# Draw the graph
nx.draw(G, with_labels=True)

# Show the plot
plt.show()
```
